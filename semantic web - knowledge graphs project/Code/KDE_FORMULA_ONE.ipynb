# Namespaces

```python
g = Graph()

FOAF = Namespace("http://xmlns.com/foaf/0.1/")
dbo = Namespace("http://dbpedia.org/ontology/")
ms = Namespace("http://example.org/motor-sports/")
f1 = Namespace("http://example.org/motor-sports/formula-one/")
wiki = Namespace("http://en.wikipedia.org/wiki/")

d = Namespace("http://example.org/motor-sports/formula-one/driver/")
rc = Namespace("http://example.org/motor-sports/formula-one/races-circuits/")
ps = Namespace("http://example.org/motor-sports/formula-one/pit_stops/")
co = Namespace("http://example.org/motor-sports/formula-one/constructors/")
s_res = Namespace("http://example.org/motor-sports/formula-one/sprint_results/")
res = Namespace("http://example.org/motor-sports/formula-one/results/")
ds = Namespace("http://example.org/motor-sports/formula-one/driver_standings/")
co_res = Namespace("http://example.org/motor-sports/formula-one/constructor_results/")
lp = Namespace("http://example.org/motor-sports/formula-one/lap_times/")
co_sta = Namespace("http://example.org/motor-sports/formula-one/constructor_standings/")
q = Namespace("http://example.org/motor-sports/formula-one/qualifying/")

g.bind("FOAF",FOAF)
g.bind("dbo",dbo)
g.bind("ms",ms)
g.bind("f1",f1)
g.bind("wiki",wiki)

g.bind("d",d)
g.bind("rc",rc)
g.bind("ps",ps)
g.bind("co",co)
g.bind("s_res",s_res)
g.bind("res",res)
g.bind("ds",ds)
g.bind("co_res",co_res)
g.bind("lp",lp)
g.bind("co_sta",co_sta)
g.bind("q",q)
```

# G1
## Drivers

```python
d_labels = pd.read_csv(dir+"drivers.csv").columns

#0.       1.        2.       3.    4.      5.     6.    7.        8
#driverId,driverRef,number,code,forename,surname,dob,nationality,url
#done.    done      done.  done done.    done.  done.  done     done

#Classes
g.add((f1.Driver, RDF.type, OWL.Class))
g.add((ms.Racers, RDF.type, OWL.Class))
g.add((dbo.Person, RDF.type, OWL.Class))
g.add((f1.Driver, RDFS.subClassOf, dbo.Person))
g.add((dbo.Country,RDF.type, OWL.Class))

g.add((f1.Driver, RDFS.subClassOf, ms.Racers))
#Properties
g.add((d.hasNationality, RDF.type, OWL.ObjectProperty))
g.add((d.hasNationality, RDFS.domain, dbo.Person))
g.add((d.hasNationality, RDFS.range, dbo.Country))

distinct_members_list = BNode()
g.add((f1.Driver, OWL.distinctMembers, distinct_members_list))

predicates_d=list(d_labels[1:])
#print(predicates)
remove_d = ["url","dob","forename","surname"]
#remove = [resultId]
for i in remove_d:
    predicates_d.remove(i)

for predicate in predicates_d:
    subject = "d."+predicate
    g.add((URIRef(subject), RDF.type, OWL.ObjectProperty))
    g.add((URIRef(subject), RDFS.domain, f1.Driver))

for i in drivers:
    #id
    g.add((d[str(i[0])],RDF.type,f1.Driver))
    g.add((distinct_members_list, RDF.first, d[str(i[0])]))

    #Literals
    for predicate in predicates_d:
        #driverRef, number, code, nationality
        if i[list(d_labels).index(predicate)] != r"\N":
            g.add((d[str(i[0])],d[predicate],Literal(str(i[list(d_labels).index(predicate)]))))

    #firstName
    g.add((d[str(i[0])],FOAF.firstName,Literal(str(i[4]))))

    #surname
    g.add((d[str(i[0])],FOAF.surname,Literal(str(i[5]))))

    #dob
    g.add((d[str(i[0])],FOAF.dob,Literal(str(i[6]),datatype=XSD.date)))

    #url
    #to avoid having the complete link, dbo gives the link http://en.wikipedia.org/wiki/
    position = str(i[8]).rfind('/')
    url = str(i[8])[position+1:]
    g.add((d[str(i[0])],RDFS.seeAlso,wiki[url]))

g.add((distinct_members_list, RDF.rest, RDF.nil))
#print(g.serialize( format="turtle"))
```

## Lap Times

We start by sorting the data into groups based on the new_ID = raceID + driverID, so that its easier to create the rdf sequences

```python
grouped_pos = {}
grouped_ms ={}

for datapoint in lap_times:
    ids = str(datapoint[0])+'_'+str(datapoint[1])
    pos = datapoint[3]
    ms = datapoint[5]
    if ids not in grouped_pos:
        grouped_pos[ids] = []
        grouped_ms[ids] = []

    grouped_pos[ids].append(pos)
    grouped_ms[ids].append(ms)
```

```python
g.bind("lp",lp)

#0.       1.      2   3.       4.  5
#raceId,driverId,lap,position,time,milliseconds

#Classes
g.add((f1.LapTime, RDF.type, OWL.Class))

#Properties
g.add((lp.hasRaceID, RDF.type, OWL.ObjectProperty))
g.add((lp.hasRaceID, RDFS.domain, f1.LapTime))
g.add((lp.hasRaceID, RDFS.range, f1.F1RaceEvents))

g.add((lp.hasDriverID, RDF.type, OWL.ObjectProperty))
g.add((lp.hasDriverID, RDFS.domain, f1.LapTime))
g.add((lp.hasDriverID, RDFS.range, f1.Driver))

g.add((lp.position, RDF.type, OWL.ObjectProperty))
g.add((lp.position, RDFS.domain, f1.LapTime))

g.add((lp.msec, RDF.type, OWL.ObjectProperty))
g.add((lp.msec, RDFS.domain, f1.LapTime))


for i,j in zip(grouped_pos,grouped_ms):
    g.add((lp[str(i)],RDF.type,f1.LapTime))

    position = i.rfind('_')
    #raceID = i[:position]
    #driverID =
    g.add((lp[str(i)],lp.hasRaceID,rc[str(i[:position])]))
    g.add((lp[str(i)],lp.hasDriverID,d[str(i[position+1:])]))

    #position
    b1 = BNode()
    g.add((lp[str(i)], lp.position, b1))
    g.add((b1, RDF.type, RDF.Seq))
    for x in range(1,len(grouped_pos[i])):
        predicate1 = str("RDF._"+str(x))
        g.add((b1, URIRef(predicate1), Literal(grouped_pos[i][x-1])))

    b2 = BNode()
    g.add((lp[str(j)], lp.msec, b2))
    g.add((b2, RDF.type, RDF.Seq))
    for y in range(1,len(grouped_ms[j])):
        predicate2 = str("RDF._"+str(y))
        g.add((b2, URIRef(predicate2), Literal(grouped_ms[j][y-1])))

#print(g.serialize(format="turtle"))
```
